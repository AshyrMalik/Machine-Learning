## which one method youshould use
* we can use pyplot for ploting something quick
* when ploting some thing advance use OOM (Obhject oriented method)


```python
over_50= heart_disease[heart_disease["age"]>50]
over_50.head()
```

```python
#pyplot method
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target",
             figsize=(10,6)
);
```

```python
#OOM 
fig, ax =plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target",
             ax=ax
);
#ax.set_xlim(45,75); used to set the values of axes
```

```python
# OO Method from scratch

fig, ax =plt.subplots(figsize=(10,6))

scatter =ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"],
                   )

# Customise the plot 
ax.set(title="Heart disease and Cholestrol",
       ylabel="Chol",
       xlabel="Age"
    
);
#customising legend
ax.legend(*scatter.legend_elements(),title="Target")

# adding a horizontal line 
ax.axhline(over_50["chol"].mean(),
          linestyle="--");
```

```python
over_50
```

```python
#Creating canves for thal,chol,heart disease
fig, (ax0,ax1)=plt.subplots(nrows=2,
                           ncols=1,
                           figsize=(10,10),
                           sharex=True)
#Adding data to ax0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"]);
#customisng ax0
ax0.set(title="Heart Disease and Cholestrol",
      ylabel="Cholestrol",
      )
ax0.legend(*scatter.legend_elements(),title="Target")

#adding horizontal line to ax0

ax0.axhline(over_50["chol"].mean(),
           linestyle="--");

# Age and thalach
#adding data
scatter==ax1.scatter(x=over_50["age"],
                    y=over_50["thalach"],
                    c=over_50["target"])
#Customisng plot
ax1.set(title="Heart Disease and Max heart rate",
       ylabel="Thalach",
       xlabel="Age")
ax1.legend(*scatter.legend_elements(),title="Target")

#horizonal line
ax1.axhline(over_50["thalach"].mean(),
            linestyle="--");


fig.suptitle("Heart Disease Analysis",size=16,weight= 800);
```

```python
#setting the style
plt.style.use('seaborn-v0_8-whitegrid')

# OO Method from scratch
fig, ax =plt.subplots(figsize=(10,6))

scatter =ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"],
                    cmap="summer"
                   )

# Customise the plot 
ax.set(title="Heart disease and Cholestrol",
       ylabel="Chol",
       xlabel="Age"
    
);
#customising legend
ax.legend(*scatter.legend_elements(),title="Target")

# adding a horizontal line 
ax.axhline(over_50["chol"].mean(),
          linestyle="--",
          color="r");
```

```python
#Creating canves for thal,chol,heart disease
fig, (ax0,ax1)=plt.subplots(nrows=2,
                           ncols=1,
                           figsize=(10,10),
                           sharex=True)
#Adding data to ax0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap="summer");
#customisng ax0
ax0.set(title="Heart Disease and Cholestrol",
      ylabel="Cholestrol",
      xlim=([50,80])
      )
ax0.legend(*scatter.legend_elements(),title="Target")

#adding horizontal line to ax0

ax0.axhline(over_50["chol"].mean(),
           linestyle="--",
            c="r");

# Age and thalach
#adding data
scatter==ax1.scatter(x=over_50["age"],
                    y=over_50["thalach"],
                    c=over_50["target"],
                    cmap="summer")
#Customisng plot
ax1.set(title="Heart Disease and Max heart rate",
       ylabel="Max heart rate",
       xlabel="Age",
       ylim=([60,200]))
ax1.legend(*scatter.legend_elements(),title="Target")

#horizonal line
ax1.axhline(over_50["thalach"].mean(),
            linestyle="--",
           c="r");


fig.suptitle("Heart Disease Analysis",size=16,weight= 800);
```
